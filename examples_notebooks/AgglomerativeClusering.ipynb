# Agglomerative Clustering Example Notebook

This notebook tests the implementation of the Agglomerative clustering algorithm implemented in KTorch

```python
%cd ..
from algorithms.cluster.AgglomerativeClustering import AgglomerativeClustering
import numpy as np 
import matplotlib.pyplot as plt
```

### Simple Example

```python
# Create a dataset
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=100, centers=3, n_features=2, random_state=42)
```

```python
# Plotting the data
plt.scatter(X[:, 0], X[:, 1])
plt.title("Blobs")
plt.show()
```

```python
# Fit the model
model = AgglomerativeClustering(n_clusters=3, linkage='average')
model.fit(X)
```

```python
set(model.labels)
```

```python
# Plotting the clusters
plt.scatter(X[:, 0], X[:, 1], c=model.labels)
plt.title("Clusters")
plt.show()
```

### More complex Example 
to view the differences between the different linkages

```python
# Create a complex dataset to view the differences between the linkage methods
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=100, noise=0.1, random_state=42)

# Plotting the data
plt.scatter(X[:, 0], X[:, 1])
plt.title("Moons")
plt.show()
```

```python
# Fit the average linkage model
model = AgglomerativeClustering(n_clusters=2, linkage='average')
labels = model.fit_predict(X)

# Plotting the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.title("Clusters")
plt.show()
```

```python
# Fit the complete linkage model
model = AgglomerativeClustering(n_clusters=2, linkage='complete')
labels = model.fit_predict(X)

# Plotting the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.title("Clusters")
plt.show()
```

It seems like that the complete linkage was better in detecting the tow moons as fewer purple data points on the top moon

```python
# Fit the single linkage model
model = AgglomerativeClustering(n_clusters=2, linkage='single')
labels = model.fit_predict(X)

# Plotting the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.title("Clusters")
plt.show()
```

It seems like the single linkage was able to detect the moons successfully.
