# KMedians Example Notebook

This notebook tests the implementation of the KMedians clustering algorithm implemented in KTorch

```python
# imports
%cd ..
from algorithms.cluster.KMedians import KMedians
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)
```

```python
# Getting the data
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=300, centers=3, n_features=2, random_state=42)
```

```python
# Plotting the data
plt.scatter(X[:, 0], X[:, 1])
plt.title("Blobs")
plt.show()
```

```python
# Initiate and fit the kmedians 
kmedians = KMedians(n_clusters=3, max_iter=400, init='random')
kmedians.fit(X)
```

```python
# Plotting the data with the clusters
labels = kmedians.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(kmedians.centroids[:, 0], kmedians.centroids[:, 1], c='red', marker='x')
plt.title("Blobs with clusters")
plt.show()
```

A problem appears because of the random initialization of 

```python
# Initiate and fit the kmedians with kmedians++
kmedians = KMedians(n_clusters=3, max_iter=400, init='kmedians++')
kmedians.fit(X)
```

```python
# Plotting the data with the clusters
labels = kmedians.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(kmedians.centroids[:, 0], kmedians.centroids[:, 1], c='red', marker='x')
plt.title("Blobs with clusters")
plt.show()
```

`KMedians++` fixed this problem
